# Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv("Data.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
# Filter the DataFrame for Ronaldo Goals 
df_Ronaldo = df[['Goals', 'Year']][df['Player Names'] == 'Cristiano Ronaldo']
df_Ronaldo.head()
```

# Data Visualization

```python
# Plot the bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Goals', data=df_Ronaldo)
plt.title('Goals Scored by Ronaldo Over the Years')
plt.xlabel('Year')
plt.ylabel('Goals')
plt.show()
```

```python
# Get top scorers that play in Real Madrid in 2016
df_madrid = df_madrid = df[(df['Club'] == '(RMA)') & (df['Year'] == 2016)]
df_madrid.head()
```

```python
# Draw a pie chart
plt.figure(figsize=(10, 6))
plt.pie(df_madrid['Goals'], labels=df_madrid['Player Names'], autopct='%1.1f%%', startangle=140)
plt.title('Goals Scored by Real Madrid Players in 2016')
plt.show()
```

```python
# Compare Between Lionel Messi and Cristiano Ronaldo Goals
players = df[['Player Names','Goals','Year']][(df['Player Names']=='Cristiano Ronaldo') | (df['Player Names']=='Lionel Messi')]
players
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(x='Player Names', y='Goals', hue='Year', data=players)
```

```python
df_laliga = df[df['League'] == 'La Liga']
df_laliga.head()
```

```python
# Sort by goals and get top 10 scorers
df_top_scorers = df_laliga.sort_values(by='Goals', ascending=False).head(10)
df_top_scorers
```

```python
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df_top_scorers['Goals'], bins=5, edgecolor='black')
plt.title('Distribution of Goals Scored by Top 10 Players in La Liga')
plt.xlabel('Number of Goals')
plt.ylabel('Frequency')
plt.show()
```
